```python
import os
from rdkit import Chem
from rdkit.Chem import Draw, AllChem
from rdkit.Chem.Draw import rdMolDraw2D

def smiles_to_svg(smi, name=None, addH=False, bw=False, size=(400, 400), out_dir="svg_mols"):
    '''Render molecule from SMILES as SVG and save to out_dir.'''
    mol = Chem.MolFromSmiles(smi)
    if mol is None:
        print(f"Invalid SMILES: {smi}")
        return
    if addH:
        mol = Chem.AddHs(mol)
    AllChem.Compute2DCoords(mol)

    drawer = rdMolDraw2D.MolDraw2DSVG(*size)
    if bw:
        opts = drawer.drawOptions()
        opts.updateAtomPalette({k: (0, 0, 0, 1) for k in range(119)})
    drawer.DrawMolecule(mol)
    drawer.FinishDrawing()

    os.makedirs(out_dir, exist_ok=True)
    filename = (name or smi).replace("/", "_").replace("\\", "_") + ".svg"
    path = os.path.join(out_dir, filename)
    with open(path, "w") as f:
        f.write(drawer.GetDrawingText())
    print(f"Saved: {path}")

def batch_svg(molecules, **kwargs):
    '''Save SVGs for a list of (smiles, name) tuples or plain SMILES strings.'''
    for entry in molecules:
        if isinstance(entry, tuple):
            smiles_to_svg(entry[0], name=entry[1], **kwargs)
        else:
            smiles_to_svg(entry, **kwargs)
```

```python
molecules = [
    ("CC(C#N)N", "2-aminopropionitrile"),
]

batch_svg(molecules, bw=True)
```
